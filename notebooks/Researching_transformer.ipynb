```python
import torch 
import torch.nn as nn
```

```python
# VOCAB_SIZE:int = 32000
# D_MODEL:int = 3 # each token will be divided into 3 dimensional vector
# MAX_SEQ_LEN:int = 128

VOCAB_SIZE:int = 32000
D_MODEL:int = 512 # each token will be divided into 3 dimensional vector
MAX_SEQ_LEN:int = 128

class Encoder(nn.Module) :
    def __init__(self): 
        super().__init__()
        self.embedding = nn.Embedding(VOCAB_SIZE, D_MODEL)
        self.ll1 = nn.Linear(D_MODEL, D_MODEL)

    def forward(self, x): 
        # x = [batch, max_sequence_len]
        x = self.embedding(x) # outputs -> [batch, max_sequence_len, d_model]
        x = self.ll1(x) # outputs -> [batch, max_sequence_len, d_model]
        return x

class Decoder(nn.Module) :
    def __init__(self):
        super().__init__()
    def forward(self):pass
```

```python
low = 0
high = VOCAB_SIZE

# Generate the 2x128 tensor
random_tensor = torch.randint(low, high, (2, 5))
encoder = Encoder()
print(f"Random Tensor: {random_tensor.tolist()}")

enc_output = encoder.embedding(random_tensor)
print(f'Embedding Output: {enc_output.tolist()}')
print(f"Embedding table: {encoder.embedding.weight}")
print(f"Embedding output shape: {enc_output.shape}")

ll1_output = encoder.ll1(enc_output)
print(f"Linear Layer output: {ll1_output.tolist()}")
print(f"Linear Layer output shape: {ll1_output.shape}")
```
